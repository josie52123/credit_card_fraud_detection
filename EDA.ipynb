Spearman's Rank Correlation Coefficient (ρ):

Evaluates the adequacy of describing the relationship between two variables using a monotonic function.
Suitable for both continuous and ordinal data.
Does not assume normal distribution.

```python
import scipy.stats

plot_columns=['trans_date_trans_time','cc_num', 'amt','zip','lat', 'long','city_pop','dob','unix_time','merch_lat','merch_long','is_fraud']
subset_df=df[plot_columns]

#Spearman's Rank Correlation Coefficient
spearman_corr=subset_df.corr(method='spearman')
print(spearman_corr)
```

```python
sns.set(rc={"figure.figsize":(16, 8)})
sns.heatmap(spearman_corr, annot=True, annot_kws={'size': 10}, cmap='coolwarm')
plt.show()
```

Kendall's Tau (τ):

Similar to Spearman's rank correlation, Kendall's tau is a non-parametric measure of association.
Assesses the strength and direction of a monotonic relationship.
Often used when dealing with ordinal data.
Less sensitive to outliers compared to Pearson's correlation.

```python
# Kendall's Tau
kendall_tau_corr = subset_df.corr(method='kendall')
print(kendall_tau_corr)
```

```python
sns.set(rc={"figure.figsize":(16, 8)})
sns.heatmap(kendall_tau_corr, annot=True, annot_kws={'size': 10}, cmap='coolwarm')
plt.show()
```
